# Brownian Walk

Randomly Generate some numbers over time (normal distribution), 
use the numpy cumulative sum function to add the path up, and plot using matplot lib and pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(345345)
time_series = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020',periods=1000))
time_series.head()
time_series = time_series.cumsum()
time_series.head()


time_series.plot()
plt.title('Brownian Walk over time')
plt.show()
```
