# Brownian Walk

Randomly Generate some numbers over time (normal distribution),
use the numpy cumulative sum function to add the path up, and plot using matplot lib and pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_walk(start_date='1/1/2020', num_steps=1_000, random_seed=345345):
  np.random.seed(random_seed)
  time_series = pd.Series(np.random.randn(num_steps), index=pd.date_range(start_date, periods=num_steps))
  time_series.head()
  time_series = time_series.cumsum()
  time_series.head()

  time_series.plot()
  plt.title('Brownian Walk over time')
  plt.show()

plot_walk()
```

```python
plot_walk(random_seed=1231)
```

```python
plot_walk(num_steps=10_000)
```
